# **Read and Prepare Data**

```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client.youtube
search_collection = db.search
videos_collection = db.videos
```

```python
searchs = search_collection.find()
videos = videos_collection.find()
```

```python
import pandas as pd
from numpy import nan
```

```python
sdf = pd.json_normalize(searchs)
sdf.drop_duplicates(subset=['id.videoId'], inplace=True)
sdf
```

```python
vdf = pd.json_normalize(videos)
vdf.drop_duplicates(subset=['id'], inplace=True)
vdf
```

# **Merge Searchs and Details**

```python
df = pd.merge(sdf, vdf, left_on=['id.videoId'], right_on=['id'], how='inner', copy=True)

df
```

# **Extract Words from Titles**

```python
df["tags"] = df.apply(lambda i: ' '.join(i['snippet.tags']) if isinstance(i['snippet.tags'], list) else '', axis=1)
```

```python
tokens = ''.join(map(lambda i: i.lower() if i.isalpha() else ' ',' '.join(map(str, df['snippet.title_x'])))).split()

with open('../persian-stop-words.txt', 'r', encoding='utf-8') as fp:
    stopwords = fp.read().split('\n')

clean_tokens = list(filter(lambda i: i not in stopwords, tokens))
```

```python
exclude = [
    'Ù…ÙˆØ²ÛŒÚ©',
    'Ø¢Ù‡Ù†Ú¯',
    'Ù…ÙˆØ³ÛŒÙ‚ÛŒ',
    'Ø§Ù‡Ù†Ú¯',
    'Ù…ÙˆØ³ÙŠÙ‚ÛŒ',
    'ØªØ±Ø§Ù†Ù‡',
    'Ú©Ù„ÛŒÙ¾',
    'ÙˆÛŒØ¯ÛŒÙˆ',
    'video',
    'music',
    'youtube',
    'song',
]
```

```python
from wordcloud import WordCloud, STOPWORDS
from matplotlib import pyplot as plt
from bidi.algorithm import get_display
import arabic_reshaper

# text = ' '.join(map(str, df['tags']))
text = get_display(arabic_reshaper.reshape(' '.join(clean_tokens).encode().decode('utf-8')))
# 

stopwords += exclude
for stopword in stopwords:
    STOPWORDS.add(get_display(arabic_reshaper.reshape(stopword)))

wordcloud = WordCloud(
    font_path='../utils/Vazirmatn-Regular.ttf',
    background_color="white",
    width=1000,
    height=860,
    max_words=1000,
    max_font_size=200,
    # min_font_size=15,
    stopwords=STOPWORDS,
    margin=10,
    random_state=42,
).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from collections import Counter
from matplotlib import pyplot as plt

filtered_tokens = list(filter(lambda i: i not in exclude, clean_tokens))

clean_tokens_counter = Counter(clean_tokens)
clean_tokens_frequency = clean_tokens_counter.most_common(30)

X = list(map(lambda i: get_display(arabic_reshaper.reshape(i[0])), clean_tokens_frequency))
y = list(map(lambda i: i[1], clean_tokens_frequency))


fig, ax = plt.subplots(figsize=(6, 8))

ax.barh(X, y, align='center')
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.invert_yaxis()
ax.set_title("Most Frequent Words", fontweight="bold")

plt.show() 
```

# **Most Rated Videos**

```python
df['statistics.viewCount'] = pd.to_numeric(df['statistics.viewCount'], errors='coerce')
df['statistics.likeCount'] = pd.to_numeric(df['statistics.likeCount'], errors='coerce')
df['statistics.commentCount'] = pd.to_numeric(df['statistics.commentCount'], errors='coerce')
```

```python
# filter persian videos

pattern = r"^(?=.*[Ø¶ØµØ«Ù‚ÙØºØ¹Ù‡Ø®Ø­Ø¬Ú†Ø´Ø³ÛŒØ¨Ù„Ø§ØªÙ†Ù…Ú©Ú¯Ù¾Ø¸Ø·Ø²Ø±Ø°Ø¯Ø¦Ùˆ])(?!.*[ÙŠØ©]).*$"

dff = df[
    df['snippet.defaultLanguage'].isin(['fa', 'fa-IR', 'fa-AF']) | 
    df['snippet.title_y'].str.contains(pattern) |
    df['snippet.description_y'].str.contains(pattern)
]

keys = ["id", "snippet.publishedAt_y", "snippet.title_y", "snippet.channelTitle_x", "snippet.categoryId", "statistics.viewCount", "statistics.likeCount", "statistics.commentCount"]
```

```python
# Top Views

dffl = dff.sort_values(by='statistics.viewCount', ascending=False)
dffl[keys]
```

```python
# Top Likes

dffv = dff.sort_values(by='statistics.likeCount', ascending=False)
dffv[keys]
```

```python
# Top Comments

dffc = dff.sort_values(by='statistics.commentCount', ascending=False)
dffc[keys]
```

# **Most Rated Channels**

```python
dfch = dff.groupby("snippet.channelId_x").agg({ "snippet.channelTitle_x": "first", 'statistics.viewCount': 'sum', 'statistics.likeCount': 'sum', 'statistics.commentCount': 'sum'})
dfch.sort_values(by='statistics.viewCount', ascending=False)
```
